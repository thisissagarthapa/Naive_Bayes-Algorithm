# Naive Bayes Algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("large_health_data.csv")
```

```python
dataset.head(4)
```

Encoding the smoking status due to categorical data


```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
le.fit_transform(dataset["Smoking_Status"])
```

```python
dataset["Smoking_Status"]=le.fit_transform(dataset["Smoking_Status"])
```

```python
sns.scatterplot(x="Age",y="Smoking_Status",hue="Lung_Cancer",data=dataset)
plt.show()
```

```python
sns.kdeplot(data=dataset["Smoking_Status"])
plt.show()
```

```python
dataset.drop(columns=["Patient_ID"],inplace=True)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Lung_Cancer"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100
```

```python
gnb.predict([[60,1]])
```
